```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, 10, 2)
y = x*10
plt.plot(x, y)
f1 = {'family' : "cambria", 'color' : "red", 'size' : 15}
f2 = {'family' : "cambria", 'color' : "green", 'size' : 20}
plt.title("Multiples of 10", fontdict = f1)
plt.xlabel(" X - Values")
plt.ylabel("X Multiply 10", fontdict = f2)
plt.show()
```
