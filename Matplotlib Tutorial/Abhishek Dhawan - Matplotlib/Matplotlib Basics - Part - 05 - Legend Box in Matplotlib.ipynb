```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, 5, 1)
y1 = x ** 2
y2 = x ** 3
plt.plot(x, y1, label = "SQUARES")
plt.plot(x, y2, label = "CUBES")
plt.legend(loc = "upper left", framealpha = 1, facecolor = "yellow", edgecolor = "black", fancybox = True, shadow = True, fontsize = 20)

plt.show()
```
