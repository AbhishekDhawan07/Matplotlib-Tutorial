```python
from matplotlib import pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [10, 50, 100, 80, 60, 90]
color = ["red", "green","blue", "black", "brown", "orange"]

plt.scatter(x, y, c = color)
plt.show()
```

```python
from matplotlib import pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [10, 50, 100, 80, 60, 90]

size = [10, 100, 20, 90, 50, 70]
plt.scatter(x, y, color = "black", s = size, alpha = 1, cmap = "virdis")
plt.colorbar()
plt.show()
```
