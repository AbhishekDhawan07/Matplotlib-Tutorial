```python
from matplotlib import pyplot as plt

years = ["2017", "2018", "2019", "2020"]
placements = [150, 275, 325, 300]


plt.title("Placement Comparision")
plt.bar(years, placements, width = 0.4)

plt.show()
```

```python
from matplotlib import pyplot as plt

years = ["2017", "2018", "2019", "2020"]
cse = [75, 80, 85, 90]
ece = [46, 58, 60, 80]
it = [100, 90, 65, 70]

plt.title("Placement Comparision")
plt.bar(years, cse, width = 0.4)
plt.bar(years, ece, width = 0.4)
plt.bar(years, it, width = 0.4)

plt.show()
```
